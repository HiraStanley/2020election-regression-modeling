```python
df.head()
```

```python
df['turnout_difference'] = df['total_votes20']-df['total_votes16']
```

```python
df['case_pct'] = df['cases']/df['TotalPop']
```

```python
df['death_pct'] = df['deaths']/df['cases']
```

```python
df.head()
```

```python
df['Men_pct'] = df['Men']/df['TotalPop']
```

```python
df_continental = df[~df['state'].isin(['AK', 'HI'])]
```

```python
sns.scatterplot(data=df_continental, x='long', y='lat', hue='percentage16_Donald_Trump', size='total_votes16', palette = 'coolwarm')
plt.legend(bbox_to_anchor=(1.8,1),loc='upper right')
plt.title('2016 Election Results')
```

```python
sns.scatterplot(data=df_continental, x='long', y='lat', hue='percentage20_Donald_Trump', size='total_votes20', palette = 'coolwarm')
plt.legend(bbox_to_anchor=(1.8,1),loc='upper right')
plt.title('2020 Election Results')
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df['2016_county_winner'] = df.apply(lambda row: 'R' if row['percentage16_Donald_Trump'] > row['percentage16_Hillary_Clinton'] else 'D', axis = 1)
```

```python
df['2020_county_winner'] = df.apply(lambda row: 'R' if row['percentage20_Donald_Trump'] > row['percentage20_Joe_Biden'] else 'D', axis = 1)
```

```python
grouped_df = df.groupby('state')[['percentage16_Donald_Trump', 'percentage16_Hillary_Clinton']].mean().reset_index()
```

```python
custom_colors = ['#FF5733', '#1E90FF'] 
grouped_df.set_index('state').plot(kind='bar', stacked='True', color=custom_colors)

plt.xlabel('State')
plt.ylabel('Percentage')
plt.title('2016 Election Results by State')

plt.show()
```

```python
grouped_df_20 = df.groupby('state')[['percentage20_Donald_Trump', 'percentage20_Joe_Biden']].mean().reset_index()
```

```python
custom_colors = ['#FF5733', '#1E90FF'] 
grouped_df_20.set_index('state').plot(kind='bar', stacked='True', color=custom_colors)

plt.xlabel('State')
plt.ylabel('Percentage')
plt.title('2020 Election Results by State')

plt.show()
```
